```python
import yfinance as yf
```

```python
start_date = datetime.datetime(2021,1,17)
end_date = datetime.datetime.today()
```

```python
print(start_date)
print(end_date)
```

```python
stock_price = pd.DataFrame()
type(stock_price)
```

Assigning the pandas dataframe type to the variable so we can do different operations with the dataframe.

```python
df = yf.download('AKA', start = start_date, end = end_date)
df
```

These headings help us gauge the performance of stock (AKA) as it moves through each day. From the start to end date we indicated.

```python
comp_energy['Symbol'].unique()
```

```python
for ticker in comp_energy['Symbol'].unique():
  df = yf.download(ticker, start = start_date, end = end_date)
  df['Symbol'] = ticker

  stock_price = stock_price.append(df)
  print(df.head())
```

for every variable in this list, assign it to the variable ticker. 'Ticker' in the first iteration contains APA and will execute the df function. After thats done, the algo adds another colum and will replace the 'ticker' with the specific ticker symbol, in the first iteration, AKA. Afterthat, we append our stock price variable with whatever the df function we written does ie: see the open , high ,low of each stock.

```python
stock_price[(stock_price['Symbol']=='APA')]
```

```python
APA = stock_price[(stock_price['Symbol'] == 'APA')]
```

```python
import plotly.graph_objects as go

import pandas as pd

df = APA

fig = go.Figure(data=[go.Candlestick(x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])])

fig.show()
```
