```python
#training the machine learning model. 
#first, we figure out what to predict, the target (eg. the maximum temperature tommorow)
#shift(-1) will pull every row back 1 position, ensuring we predict tommorow's temp
```

```python
core_weather['target'] = core_weather.shift(-1)['temp_max']
```

```python
core_weather
```

```python
core_weather = core_weather.iloc[:-1,:].copy()
#deleting the row with last value because the last row won't have a next day
```

```python
core_weather
```

```python
#using ridge regression (regression that minimises overfitting)

from sklearn.linear_model import Ridge
reg = Ridge(alpha=.1)
```

```python
#Now, we will identify what columns we will use to make our predictions
predictors = ['precip', 'temp_max', 'temp_min']
```

```python
# make a training and testing dataset. Make sure to make training data historical and testing data more present.
train = core_weather.loc[:'2021-01-10']
test = core_weather.loc['2021-01-01':]
```

```python
reg.fit(train[predictors], train['target'])
```

```python
predictions = reg.predict(test[predictors])
```

```python
#check the model's accuracy. On average, the model's prediction is 3.4 degrees off
from sklearn.metrics import mean_absolute_error

mean_absolute_error(test['target'], predictions)

```

```python
#evaluating our model
combined = pd.concat([test['target'], pd.Series(predictions, index=test.index)], axis=1)
#renaming the columns
combined.columns = ['actual', 'predictions']
```

```python
'''Creating a function we can call for predictions to enable quick iteration.def create_predictions(predictors, core_weather, reg):
   1. split the dataset into train and test
   2. fit the model
   3. make predictions
   4. error metric
   5. combine actual values with predicted values'''

    train = core_weather.loc[:'2021-01-10']
    test = core_weather.loc['2021-01-01':]
    
    reg.fit(train[predictors], train['target'])
    
    predictions = reg.predict(test[predictors])
    
    error = mean_absolute_error(test['target'], predictions)
    
    combined = pd.concat([test['target'], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ['actual', 'predictions']
    
    return error, combined
    
```

```python
#feature engineering. We want to create more features. We will add in rolling means.  This variable we will call `month_max` is a rolling thirty day average of `temp_max`

core_weather['month_max'] = core_weather['temp_max'].rolling(30).mean()
core_weather
```
